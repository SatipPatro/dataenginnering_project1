```python
import pandas as pd
import psycopg2 as psy
```

```python
#define function for creating database 
def create_database():
    conn = psy.connect("host=127.0.0.1 dbname=postgres user=postgres password=12345")
    conn.set_session(autocommit=True)
    cur = conn.cursor()
    
    #create database
    cur.execute("Drop database if exists accounts")
    cur.execute("Create database accounts")
    
    #closing connection from default database
    conn.close()
    
    #connecting to the new database created
    conn= psy.connect("host=127.0.0.1 dbname=accounts user=postgres password=12345")
    conn.set_session(autocommit=True)
    cur=conn.cursor()
    
    return cur,conn 
    
```

```python
cur,conn =create_database()
```

```python
accounts_data_tablecreate=("""CREATE TABLE IF NOT EXISTS accountsdata(
Country_Name varchar, 
Country_Code varchar, 
Series_Name varchar, 
Series_Code varchar,
YR1995 numeric,
YR2000 numeric,
YR2005 numeric,
YR2010 numeric,
YR2015 numeric)""")
cur.execute(accounts_data_tablecreate)
conn.commit()
```

```python
accounts_country_tablecreate=("""CREATE TABLE IF NOT EXISTS accountscountry(
Code varchar, 
Long_Name varchar, 
Short_Name varchar, 
Table_Name varchar,
Currency_Unit varchar
)""")
cur.execute(accounts_country_tablecreate)
conn.commit()
```

```python
accounts_series_tablecreate=("""CREATE TABLE IF NOT EXISTS accountsseries(
Code varchar, 
Long_Name varchar, 
Short_Name varchar, 
Table_Name varchar,
Currency_Unit varchar
)""")
cur.execute(accounts_series_tablecreate)
conn.commit()
```

```python
accounts_data_table_insert=("""INSERT INTO  accountsdata(
Country_Name , 
Country_Code , 
Series_Name , 
Series_Code ,
YR1995 ,
YR2000 ,
YR2005 ,
YR2010 ,
YR2015 )
values (%s,%s,%s,%s,%s,%s,%s,%s,%s)
""")
```

```python
#inserting rows from dataframe to country table through iterating
for i,row in Accountcountry_clean.iterrows():
    #print(i,list(row))
    cur.execute(accounts_country_table_insert,list(row))
conn.commit()
```
